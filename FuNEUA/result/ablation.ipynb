```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_result_100 = pd.read_csv('./ablation/06220000_100_0.5_1_40result.csv').reset_index()
df_result_200 = pd.read_csv('./ablation/06220014_200_0.5_1_40result.csv').reset_index()
df_result_300 = pd.read_csv('./ablation/06220035_300_0.5_1_40result.csv').reset_index()
df_result_400 = pd.read_csv('./ablation/06220059_400_0.5_1_40result.csv').reset_index()
df_result_500 = pd.read_csv('./ablation/06220120_500_0.5_1_40result.csv').reset_index()
df_result_600 = pd.read_csv('./ablation/06220142_600_0.5_1_40result.csv').reset_index()
df_result_700 = pd.read_csv('./ablation/06220202_700_0.5_1_40result.csv').reset_index()
df_result_800 = pd.read_csv('./ablation/06220223_800_0.5_1_40result.csv').reset_index()
df_result_900 = pd.read_csv('./ablation/06220247_900_0.5_1_40result.csv').reset_index()
df_result_1000 = pd.read_csv('./ablation/06220313_1000_0.5_1_40result.csv').reset_index()
dfs = [df_result_100, df_result_200, df_result_300, df_result_400, df_result_500, df_result_600, df_result_700, df_result_800, df_result_900, df_result_1000]
```

```python
fnn_user_1 = []
fnn_server_1 = []
fnn_user_2 = []
fnn_server_2 = []
fnn_user = []
fnn_server = []
Random_user = []
Random_server = []
Greedy_user = []
Greedy_server = []
MCF_user = []
MCF_server = []
DRoEUA_user = []
DRoEUA_server = []
for df in dfs:
    fnn_user_1.append(df['fnn1_user'].mean())
    fnn_server_1.append(df['fnn1_server'].mean())
    fnn_user_2.append(df['fnn2_user'].mean())
    fnn_server_2.append(df['fnn2_server'].mean())
    fnn_user.append(df['fnn_user'].mean())
    fnn_server.append(df['fnn_server'].mean())
    Random_user.append(df['random_user'].mean())
    Random_server.append(df['random_server'].mean())
    Greedy_user.append(df['greedy_user'].mean())
    Greedy_server.append(df['greedy_server'].mean())
    MCF_user.append(df['mcf_user'].mean())
    MCF_server.append(df['mcf_server'].mean())
    DRoEUA_user.append(df['fuzzy_user'].mean())
    DRoEUA_server.append(df['fuzzy_server'].mean())

xlabel=[100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]

fnn_user_scaled_1 = [x * 100 for x in fnn_user_1]
fnn_user_scaled_2 = [x * 100 for x in fnn_user_2]
fnn_user_scaled = [x * 100 for x in fnn_user]
Random_user_scaled = [x * 100 for x in Random_user]
Greedy_user_scaled = [x * 100 for x in Greedy_user]
MCF_user_scaled = [x * 100 for x in MCF_user]
DRoEUA_user_scaled = [x * 100 for x in DRoEUA_user]

fnn_server_scaled_1 = [x * 100 for x in fnn_server_1]
fnn_server_scaled_2 = [x * 100 for x in fnn_server_2]
fnn_server_scaled = [x * 100 for x in fnn_server]
Random_server_scaled = [x * 100 for x in Random_server]
Greedy_server_scaled = [x * 100 for x in Greedy_server]
MCF_server_scaled = [x * 100 for x in MCF_server]
DRoEUA_server_scaled = [x * 100 for x in DRoEUA_server]
print(fnn_user_scaled)
print(fnn_server_scaled)
```

```python
# 
fig, ax = plt.subplots(figsize=(8, 5))
sns.set(style="ticks", rc={"lines.linewidth": 1.5})  # 

palette = sns.color_palette("husl", 7)  # 

# 
markers = {"FuNEUA": "o", "Random": "s", "Greedy": "^", "MCF": "D", "DRoEUA": "X", "NAT-EUA": "*", "Fu-EUA": "P"}

# 
for name, data in [("NAT-EUA", fnn_user_scaled_1),("Fu-EUA", fnn_user_scaled_2),("FuNEUA", fnn_user_scaled),("Random", Random_user_scaled), 
                   ("Greedy", Greedy_user_scaled),("MCF", MCF_user_scaled), ("DRoEUA", DRoEUA_user_scaled)]:
    sns.lineplot(x=xlabel, y=data, label=name, color=palette[list(markers.keys()).index(name)], 
                 marker=markers[name], markersize=10)

# 
plt.grid(axis='both', which='major', linestyle='-', linewidth=0.8, color='gray')
plt.grid(axis='both', which='minor', linestyle=':', linewidth=0.5, color='gray')

# 
plt.suptitle("(a) User Allocation Rate", x=0.5, y=0.05)  # 
# 
plt.subplots_adjust(bottom=0.18)  # bottom


plt.xlabel("Number of Users", fontsize=12)
plt.ylabel("User Allocation Rate (%)", fontsize=12)  # y

# X
plt.xticks(xlabel)

plt.savefig('user_ablation.svg')
plt.show()
```
